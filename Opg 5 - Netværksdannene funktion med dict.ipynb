```python
import numpy as np

def make_web(n,k,kmin=0):

    # Input: n og k er ikke-negative heltal
    # Output: web er en dictionary med n nøgler.
    # Værdien af hver nøgle er en mængde, der er en delmængde af nøglerne.
    
    assert(k < n), "k skal være mindre end n (da man ikke kan linke til sig selv)"
    assert(kmin <= k), "kmin skal være mindre end eller lig med k"
    keys = list(range(n))
    web = dict()
    
    for j in keys:
        numlinks = np.random.choice(range(kmin, k + 1))
        web[j] = set(np.random.choice([l for l in keys if l != j], size=numlinks, replace=False))  # INDSÆT KODE HER - Vælg et antal links (numlinks) fra de andre sider, undgå at vælge den nuværende side (j) og sørg for, at der ikke er duplikatlinks

    return web

make_web(5, 2) #returnerer en web med 5 sider, hvor hver side har mellem 0 og 2 links til andre sider.
```

# Dokumentation: Algoritmisk bygning af netværksgraf (Opgave 5)

Dette dokument beskriver logikken bag genereringen af et tilfældigt netværk (en rettet graf) repræsenteret som en naboliste (Adjacency List). Som ingeniørstuderende er det essentielt at forstå både den grafteoretiske model og den tidskompleksitet, der ligger til grund for datastrukturen.

## 1. Opsætning og Sikkerhedsnet (Guard Clauses)
Før vi allokerer hukommelse, verificeres parametrene matematisk for at forhindre logiske fejl i grafstrukturen.
* **Ingen selvreference:** Det sikres, at det maksimale antal udgående links ($k$) er strengt mindre end det samlede antal noder ($n$). En hjemmeside kan i vores model ikke linke til sig selv, så med $n$ sider findes der maksimalt $n-1$ mulige destinationer.
* **Gyldigt interval:** Det tjekkes, at minimumsgrænsen for links ikke overstiger maksimumsgrænsen.

## 2. Initialisering af Datastrukturen
* **Univers af noder:** Vi definerer et indeks over alle hjemmesiders ID'er (fra $0$ til $n-1$).
* **Adjacency List (Dictionary):** Grafen initialiseres som et Hash Map. Dette er et bevidst algoritmisk valg, da opslagstiden for at finde en specifik nodes udgående links hermed reduceres til gennemsnitligt $\mathcal{O}(1)$.

## 3. Den Iterative Udvælgelse (Maskinrummet)
For hver enkelt node i netværket udføres to operationer for at generere tilfældige, men matematisk valide, kanter:

### Trin A: Fastlæggelse af udgående grad
Algoritmen trækker først et stokastisk heltal inden for det tilladte interval (fra minimum til maksimum). Dette definerer den specifikke nodes udgående grad ($deg^+$).

### Trin B: Konstruktion af kanter
For at skabe forbindelserne udføres fire operationer simultant:
1. **Filtrering af selv-løkker:** Systemet danner lynhurtigt en kandidatpulje af alle tænkelige destinationer, hvor den nuværende node ekskluderes fra listen. 
2. **Stokastisk trækning:** Fra kandidatpuljen udvælges det præcise antal links, der blev defineret i Trin A.
3. **Forebyggelse af duplikater (Parallelle kanter):** Trækningen sker strengt *uden tilbagelægning*. Dette garanterer, at der ikke oprettes to identiske links mellem de samme to sider.
4. **Optimering via mængder (Sets):** Slutresultatet konverteres til en matematisk mængde. Senere i PageRank-beregningerne skal vi hyppigt teste for mængde-tilhørsforhold (kant-eksistens). At tjekke tilstedeværelsen i et *Set* er en $\mathcal{O}(1)$ operation. Dette er et kritisk designvalg, der forhindrer, at algoritmens køretid eksploderer eksponentielt, når netværket senere skaleres op til tusindvis af noder.
